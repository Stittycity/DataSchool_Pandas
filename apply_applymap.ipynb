```python
# How do I apply a function to pandas Series or DataFrame?
```

```python
import pandas as pd
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
train
```

```python
train.head()
```

```python
train['Sex_num'] = train.Sex.map({'female':0, 'male':1})
# created a new series, dummy variables for male and female
```

```python
train.loc[0:4, ['Sex', 'Sex_num']]
```

```python
train['Name_length'] = train.Name.apply(len)
# apply, when used as a series method, apply the function named to every string in series
```

```python
train.loc[0:4, ['Name', 'Name_length']]
```

```python
import numpy as np
```

```python
train['Fare_ceil'] = train.Fare.apply(np.ceil)
```

```python
train.loc[0:4, ['Fare', 'Fare_ceil']]
```

```python
train.Name.str.split(',').head()
# split it into a series of python lists, each list is made up of strings
# comma is separating list elements
```

```python
def get_element(my_list, position):
    return my_list[position]
# created a function called get_element
# if we pass it a list, and give it a position it will return whatever is in that position in the list
# my_list is recognized by pandas as any list you put in the funciton
```

```python
train.Name.str.split(',').apply(get_element, position=0)
# we're apply the get_element function to every list that we created prior in that line of code
```

```python
train.Name.str.split(',').apply(lambda x: x[0]).head()
```

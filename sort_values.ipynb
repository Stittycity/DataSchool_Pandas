```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies.title.sort_values()
# sorts a series (column in a DataFrame)
```

```python
movies['title'].sort_values(ascending=False)
# another version of sort_values, in descending order
```

```python
movies.head()
# showing sort doesn't change the default order of the DataFrame
```

```python
movies.sort_values(['content_rating', 'duration'])
# you can sort by multiple series/columns at a time
```
